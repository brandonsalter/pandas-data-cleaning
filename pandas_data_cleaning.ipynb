```python
url = 'https://raw.githubusercontent.com/brandonsalter/data/main/Jets_1960-2021_pre-clean.csv'
df = pd.read_csv(url) 
df
```

```python
df.info()
```

```python
df.Year.unique()
```

```python
df.loc[:,['Year']] = df.loc[:,'Year'].ffill()
df['Year'] = df['Year'].astype('int')
df
```

```python
df.Year.unique()
```

```python
df.Week.unique()
```

```python
df = df[df.Week.notnull()]
df.Week.unique()
```

```python
df = df[df.Week != 'Championship']
df = df[df.Week != 'SuperBowl']
df = df[df.Week != 'Division']
df = df[df.Week != 'Wild Card']
df = df[df.Week != 'Conf. Champ.']
df.Week.unique()
```

```python
df.Opp.unique()
```

```python
df = df[df.Opp != 'Bye Week']
df
```

```python
df = df.reset_index(drop=True)
df
```

```python
df.loc[:,['OT']] = df.loc[:,'OT'].fillna(0)
df.loc[:,['Away']] = df.loc[:,'Away'].fillna(0)
df.loc[:,['Off_TO']] = df.loc[:,'Off_TO'].fillna(0)
df.loc[:,['Def_TO']] = df.loc[:,'Def_TO'].fillna(0)
df
```

```python
df.isna().sum()
```

```python
df.loc[:,['Off_PassY']] = df.loc[:,'Off_PassY'].fillna(0)
df.loc[:,['Def_PassY']] = df.loc[:,'Def_PassY'].fillna(0)
```

```python
df.isna().sum()
```

```python
df[['Month', 'Day']] = df['Date'].str.split(' ', 1, expand=True)
df
```

```python
df = df.rename(columns={'Week': 'Week_of_Season',
          'Day': 'Day_of_Month',
           'Win/Loss': 'Win',
          })
df
```

```python
df = df.drop(['Date', 'Rec', 'Offense_Exp_Pts', 'Defense_Exp_Pts', 'Sp_Tms_Exp_Pts'], axis=1)
df
```

```python
df = df[['Week_of_Season','Month','Day_of_Month','Year','Time','Win','OT','Away','Opp','Score_Tm',
        'Score_Opp','Off_1stD','Off_TotYd','Off_PassY','Off_RushY','Off_TO','Def_1stD','Def_TotYd','Def_PassY',
         'Def_RushY','Def_TO']]
df
```

'Boston Patriots' --> 'New England Patriots'

'Dallas Texans' --> 'Kansas City Chiefs'

'San Diego Chargers' --> 'Los Angeles Chargers'

'Baltimore Colts' --> 'Indianapolis Colts'

'St. Louis Cardinals' --> 'Arizona Cardinals'

'Washington Redskins' --> 'Washington Commanders'

'St. Louis Rams' --> 'Los Angeles Rams'

 
'Tennessee Oilers' --> 'Tennessee Titans'
'Houston Oilers' --> 'Tennessee Titans'

'Los Angeles Raiders' --> 'Las Vegas Raiders' 
'Oakland Raiders' --> 'Las Vegas Raiders'

```python
df['Opp'] = df['Opp'].replace(['Boston Patriots', 'Dallas Texans', 'San Diego Chargers', 
                               'Baltimore Colts', 'St. Louis Cardinals', 'Washington Redskins', 
                               'St. Louis Rams'
                              ], 
                              ['New England Patriots', 'Kansas City Chiefs', 'Los Angeles Chargers', 
                               'Indianapolis Colts', 'Arizona Cardinals', 'Washington Commanders', 
                               'Los Angeles Rams'])

df['Opp'] = df['Opp'].replace(['Tennessee Oilers','Houston Oilers'
                              ], 
                              'Tennessee Titans')

df['Opp'] = df['Opp'].replace(['Los Angeles Raiders','Oakland Raiders'
                              ], 
                              'Las Vegas Raiders')

df.Opp.unique()
```

```python
len(df.Opp.unique()) # 32 teams minus the Jets
```

```python
df['OT'] = df['OT'].replace(['OT'], 1)
df.OT.unique()
```

```python
df['Away'] = df['Away'].replace(['@'], 1)
df.Away.unique()
```

```python
df['Win'] = df['Win'].replace(['W','T','L'], [1,0,-1])
df.Win.unique()
```

```python
df
```

```python
df = df.drop(['Time'], axis=1) # do something with time
df
```

```python
df.isna().sum()
```

```python
month_mapping = {'January':1, 'February':2, 'March':3, 'April':4,
                 'May':5, 'June':6, 'July':7, 'August':8,
                 'September':9, 'October':10, 'November':11, 'December':12}
df = df.assign(
    Month = lambda x: x['Month'].map(month_mapping)
    )
df
```

```python
df.Month.unique()
```

```python
df = pd.get_dummies(df[['Week_of_Season','Month','Day_of_Month',
                        'Year','Win','OT','Away','Opp',
                        'Score_Tm','Score_Opp','Off_1stD',
                        'Off_TotYd','Off_PassY','Off_RushY',
                        'Off_TO','Def_1stD','Def_TotYd',
                        'Def_PassY','Def_RushY','Def_TO']],
              columns = ['Opp'])

df
```

```python
df = df.rename(columns={'Opp_Arizona Cardinals':'Cardinals', 'Opp_Atlanta Falcons':'Falcons', 'Opp_Baltimore Ravens':'Ravens', 
           'Opp_Buffalo Bills':'Bills', 'Opp_Carolina Panthers':'Panthers', 'Opp_Chicago Bears':'Bears', 
           'Opp_Cincinnati Bengals':'Bengals', 'Opp_Cleveland Browns':'Browns', 'Opp_Dallas Cowboys':'Cowboys', 
           'Opp_Denver Broncos':'Broncos', 'Opp_Detroit Lions':'Lions', 'Opp_Green Bay Packers':'Packers',
           'Opp_Houston Texans':'Texans', 'Opp_Indianapolis Colts':'Colts', 'Opp_Jacksonville Jaguars':'Jaguars',
           'Opp_Kansas City Chiefs':'Chiefs', 'Opp_Las Vegas Raiders':'Raiders', 'Opp_Los Angeles Chargers':'Chargers', 
           'Opp_Los Angeles Rams':'Rams', 'Opp_Miami Dolphins':'Dolphins', 'Opp_Minnesota Vikings':'Vikings', 
           'Opp_New England Patriots':'Patriots', 'Opp_New Orleans Saints':'Saints', 'Opp_New York Giants':'Giants', 
           'Opp_Philadelphia Eagles':'Eagles', 'Opp_Pittsburgh Steelers':'Steelers', 'Opp_San Francisco 49ers':'49ers',
           'Opp_Seattle Seahawks':'Seahawks', 'Opp_Tampa Bay Buccaneers':'Buccaneers', 'Opp_Tennessee Titans':'Titans',
           'Opp_Washington Commanders':'Commanders'}
         )
df
```
